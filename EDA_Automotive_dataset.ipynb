# **Data Understanding**

```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading dataset
olx_car = pd.read_csv('OLX_cars_dataset.csv')
olx_car.head()
```

# **Data Cleaning**

```python
# Dropping irrelevant columns
olx_car = olx_car.drop(['Ad ID', 'Car Name', 'Description', 'Images URL\'s', 'Car Page', 'Car Features'], axis = 1)
olx_car.head()
```

```python
# Checking dataset shape
olx_car.shape
```

```python
# Checking missing values
olx_car.isna().sum()
```

```python
# Checking duplicated values
olx_car.duplicated().sum()
```

```python
# Filtering rows that have duplicated values
olx_car[olx_car.duplicated(keep=False)]
```

```python
# Dropping duplicated values
olx_car = olx_car.drop_duplicates()
olx_car.count()
```

# **Descriptive Statistics**

```python
# Getting dataset information
olx_car.info()
```

```python
# Getting statistics descriptive
olx_car.describe()
```

# **Visualizations**

**Checking Outliers**

```python
# Yaer box plot
plt.figure(figsize=(5,1))
sns.boxplot(x=olx_car['Year'], fliersize=1)
plt.title('Year box plot');
```

```python
# KM's driven box plot
plt.figure(figsize=(5,1))
sns.boxplot(x=olx_car['KM\'s driven'], fliersize=1)
plt.title('KM\'s driven box plot');
```

**Finding Distributions**

```python
# Creating histogram function
def histogram_func(column_str, median_text=True, **keywargs):
    plt.figure(figsize=(4,2))
    hist = sns.histplot(x=olx_car[column_str], **keywargs)
    hist.set_xticklabels(hist.get_xticklabels(), rotation=40, ha="right")
    plt.title(f'{column_str} histogram');
```

```python
histogram_func('Make')
```

```python
histogram_func('Assembly')
```

```python
histogram_func('Car documents')
```

```python
# Creating Distribution function for numeric values
def dist_func(column_str, median_text=True, **keywargs):
    plt.figure(figsize=(5,3))
    dist = sns.distplot(olx_car[column_str], kde = True)
    dist.set_xticklabels(dist.get_xticklabels(), ha="right")
    plt.title(f'{column_str} distribution');
```

```python
dist_func('KM\'s driven')
```

```python
dist_func('Price')
```

**Finding Correlation**

```python
# Converting categorical data to nnumerical using label encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()


car_corr = olx_car
car_corr['Make'] = label_encoder.fit_transform(olx_car['Make'])
car_corr['Model'] = label_encoder.fit_transform(olx_car['Model'])
car_corr['Fuel'] = label_encoder.fit_transform(olx_car['Fuel'])
car_corr['Registration city'] = label_encoder.fit_transform(olx_car['Registration city'])
car_corr['Car documents'] = label_encoder.fit_transform(olx_car['Car documents'])
car_corr['Assembly'] = label_encoder.fit_transform(olx_car['Assembly'])
car_corr['Transmission'] = label_encoder.fit_transform(olx_car['Transmission'])
car_corr['Seller Location'] = label_encoder.fit_transform(olx_car['Seller Location'])
```

```python
# Finding correlation in heatmap
plt.figure(figsize=(10,6))
sns.heatmap(car_corr.corr(), cmap='Blues', annot=True)
plt.title('Correlation Matrix')
plt.show()
```

**Finding Relationships**

```python
# Finding relationship in scatterplot
plt.figure(figsize=(5,3))
plt.scatter(olx_car['Year'], olx_car['Price'], cmap='Blues')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Price and Year Scatterplot')
plt.show()
```

```python
# Finding relationship in scatterplot
plt.figure(figsize=(5,3))
plt.scatter(olx_car['KM\'s driven'], olx_car['Price'], cmap='Blues')
plt.xlabel('Price')
plt.ylabel('KM\'s driven')
plt.title('Price and KM\'s driven Scatterplot')
plt.show()
```
